## MULTICOLINEARITY CHECK -VIF

```python
from scipy import stats
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def cal_vif(x):
    vif=pd.DataFrame()
    vif['Variables']=x.columns
    vif['VIF_FACTOR']=[variance_inflation_factor(x.values,i)for i in range(x.shape[1])]
    return(vif)
```

```python
cal_vif(x)
```

### DROPPING OUTLET IDENTIFIER AS VIF SCORE IS HIGHER THAN THE ACCEPTABLE RANGE OF 10 

```python
x1=x.drop(['Outlet_Identifier'],axis=1)
```

### CHECKING VIF SCORE AGAIN 

```python
def cal_vif(x1):
    vif=pd.DataFrame()
    vif['Variables']=x1.columns
    vif['VIF_FACTOR']=[variance_inflation_factor(x1.values,i)for i in range(x1.shape[1])]
    return(vif)
```

```python
cal_vif(x1)
```

### ALL THE COLUMNS ARE WITHIN THE RANGE OF 10, HENCE WE CAN CONCLUDE THAT THERE IS NO MULTICOLINEARITY PRESENT IN THE DATA SET
